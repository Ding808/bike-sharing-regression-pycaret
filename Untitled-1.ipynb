```python
import pandas as pd

data_path = "day.csv"   
df = pd.read_csv(data_path)

df.head()
```

```python
df.info()
df.isna().sum()
```

```python
data = df.copy()

data.columns
```

```python
from pycaret.regression import *

s = setup(
    data = data,
    target = 'cnt',                
    session_id = 123,           
    ignore_features = ['instant', 'casual', 'registered'], 
    numeric_features = None,        
    categorical_features = None,   
    normalize = True,              
    train_size = 0.8,               
    fold = 5                        
)
```

```python
best_model = compare_models()
best_model
```

```python
plot_model(best_model, plot='residuals')
plot_model(best_model, plot='error')
plot_model(best_model, plot='feature')
```

```python
new_data = data.copy()
new_data.drop('cnt', axis=1, inplace=True)

new_data.head()
```

```python
predictions = predict_model(best_model, data=new_data)
predictions.head()
```

```python
save_model(best_model, 'bike_sharing_cnt_pipeline')
```

```python
loaded_pipeline = load_model('bike_sharing_cnt_pipeline')
loaded_pipeline
```

```python
predictions_loaded = predict_model(loaded_pipeline, data=new_data)
predictions_loaded.head()
```
